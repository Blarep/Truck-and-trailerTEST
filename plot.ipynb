```python
import matplotlib.pyplot as plt

def plot_tsp_solution(solutions, eval, filename):
    # Plot nodes and edges for each vehicle
    print(f"Evaluación: {eval}")
    colors = ["orange", "red", "green", "violet", "purple"]
    
    # Crear la figura una sola vez
    plt.figure(figsize=(10, 8))
    
    for index, solution in enumerate(solutions):
        # Dibujar la ruta
        for i in range(len(solution) - 1):
            plt.plot([solution[i][0], solution[i+1][0]], 
                    [solution[i][1], solution[i+1][1]], 
                    color=colors[index % len(colors)],
                    alpha=0.7,
                    linewidth=2)
        
        # Dibujar el depot solo una vez
        if index == 0:
            plt.scatter(solution[0][0], solution[0][1], 
                       color="red", 
                       s=100,
                       label='Depot',
                       zorder=5)
        
        # Dibujar los clientes
        normal_clients = [(x, y) for x, y, is_truck in solution if is_truck == 0]
        truck_clients = [(x, y) for x, y, is_truck in solution if is_truck == 1]
        
        if normal_clients:
            x, y = zip(*normal_clients)
            plt.scatter(x, y, 
                       color="blue", 
                       label='Regular Customers' if index == 0 else "",
                       zorder=4)
        
        if truck_clients:
            x, y = zip(*truck_clients)
            plt.scatter(x, y, 
                       color="green", 
                       label='Truck Customers' if index == 0 else "",
                       zorder=4)
    
    plt.title(f'TSP Solution\nEvaluation: {eval}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()

def getsolution(filename):
    solutions = []
    current_route = []
    eval = None
    depot_id = 2  # El depot es el nodo 2 según el ejemplo
    
    try:
        with open(filename, "r") as file:
            for line in file:
                elements = line.strip().split()
                
                if "E" in elements[0]:  # Línea de evaluación
                    eval = float(elements[1])  # Guardamos el valor de evaluación
                    continue
                    
                # Procesar las coordenadas
                points = []
                coordinates = {}  # Diccionario temporal para guardar las coordenadas
                
                # Leer los puntos de la ruta
                for point in elements:
                    point_id = int(point)
                    # Asumimos que las coordenadas serían proporcionadas o calculadas
                    # Por ahora, creamos coordenadas dummy basadas en el ID
                    x = point_id * 10
                    y = point_id * 10
                    coordinates[point_id] = (x, y, 0)  # 0 para cliente normal
                
                # Crear la ruta
                current_route = []
                for point_id in elements:
                    point_id = int(point_id)
                    current_route.append(coordinates[point_id])
                    
                    # Si volvemos al depot y ya tenemos puntos, cerramos la ruta
                    if point_id == depot_id and len(current_route) > 1:
                        solutions.append(current_route)
                        current_route = [coordinates[depot_id]]
                        
                if current_route and len(current_route) > 1:
                    solutions.append(current_route)
                    
    except FileNotFoundError:
        print(f"Error: No se encontró el archivo {filename}")
        return None, None
    except Exception as e:
        print(f"Error procesando el archivo: {str(e)}")
        return None, None
        
    return solutions, eval

# Plot the TSP solution with multiple vehicles
filename = "/home/bryangzr/TruckTrailer/Truck-and-trailer/facti_sols/chao_6.30c/SolutionC/chao_6.30c_SolutionC_run1.txt"
solutions,eval = getsolution(filename)
plot_tsp_solution(solutions,eval,filename)
```
