```python
import matplotlib.pyplot as plt

def plot_instance(filename):
    with open(filename, 'r') as file:
        # Skip the first line
        next(file)
        
        # Read deposit
        deposit_data = next(file).split()
        deposit_x = float(deposit_data[1])
        deposit_y = float(deposit_data[2])
        
        # Read client data
        clients = []
        for line in file:
            client_data = line.split()
            client_x = float(client_data[1])
            client_y = float(client_data[2])
            client_number = int(client_data[0])
            client_demand = float(client_data[3])
            truck_customer = bool(int(client_data[4]))
            clients.append((client_x, client_y, client_number, client_demand, truck_customer))
    
    # Plot deposit
    plt.scatter(deposit_x, deposit_y, color='red', label='Deposit', marker='s')
    plt.text(deposit_x, deposit_y, f'Deposit', horizontalalignment='right')
    
    # Plot clients
    for client in clients:
        client_x, client_y, client_number, client_demand, truck_customer = client
        marker = 'x' if truck_customer else 'o'
        color = 'green' if truck_customer else 'blue'
        label = f'Client {client_number}\nDemand: {client_demand}'
        plt.scatter(client_x, client_y, color=color, label=label, marker=marker)
        plt.text(client_x, client_y, f'({client_number}, {client_demand})', horizontalalignment='right')

    # Set plot title and labels
    plt.title(filename)
    plt.xlabel('X')
    plt.ylabel('Y')
    
    # Show plot
    plt.grid(True)
    plt.show()

# Call the function with the filename of the instance
plot_instance('instancesMedium/chao_8.25.dat')


```
